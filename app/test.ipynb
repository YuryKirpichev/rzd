```python
import pandas as pd
import math
import numpy as np
```

```python
df = pd.read_excel('/Users/yukirpichev/Desktop/moscow_radiology/rzd/dataset.xlsx')
df.loc[df['Наименование полигона'].isnull(), 'grouped'] = False
df.loc[df['Наименование полигона'].notnull(), 'grouped'] = True

#filling the values with previous
df = df.fillna(method='ffill')

#removing the missing values
df = df[df['grouped'] != True]


```
